```python
# // Question No. 6 \\

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Preprocess the 'Age' feature using label encoding
label_encoder_age = LabelEncoder()
df['Age_encoded'] = label_encoder_age.fit_transform(df['Age'])

# Preprocess the 'Income Level' feature using label encoding
label_encoder_income = LabelEncoder()
df['Income_encoded'] = label_encoder_income.fit_transform(df['Income Level'])

# Selecting subset of features
features = df[['Age_encoded', 'Income_encoded', 'Time Spent Online (hrs/weekday)', 'Time Spent Online (hrs/weekend)', 'Likes and Reactions', 'Click-Through Rates (CTR)', 'Conversion Rates']]

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Applying K-means clustering
kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(scaled_features)

# Adding cluster labels to the dataframe
df['Cluster'] = clusters

# Analyzing the clusters
cluster_counts = df['Cluster'].value_counts()
print("Cluster Counts:")
print(cluster_counts)

# Visualizing the clusters
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Cluster', hue='Cluster', palette='Set2', legend=False)
plt.title('Cluster Distribution')
plt.xlabel('Cluster')
plt.ylabel('Count')
plt.show()
```

```python
# // Question No. 7 \\

# Grouping the dataframe by 'Cluster'
cluster_groups = df.groupby('Cluster')

# Computing mean values of numerical features and mode for categorical features within each cluster
cluster_summary = cluster_groups.agg({
    'Age_encoded': 'mean',
    'Income_encoded': 'mean',
    'Time Spent Online (hrs/weekday)': 'mean',
    'Time Spent Online (hrs/weekend)': 'mean',
    'Likes and Reactions': 'mean',
    'Click-Through Rates (CTR)': 'mean',
    'Conversion Rates': 'mean',
    'Gender': lambda x: x.mode()[0],
    'Location': lambda x: x.mode()[0],
    'Language': lambda x: x.mode()[0],
    'Education Level': lambda x: x.mode()[0],
    'Device Usage': lambda x: x.mode()[0],
    'Top Interests': lambda x: x.mode()[0]
})

# Displaying the cluster summary
print(cluster_summary)
```

```python
# // Question No. 8 \\

# Function to assign cluster names based on characteristics observed in the data
def assign_cluster_name(cluster_row):
    # Extracting values from the cluster row
    weekend_activity = cluster_row['Time Spent Online (hrs/weekend)']
    likes_reactions = cluster_row['Likes and Reactions']
    gender = cluster_row['Gender']
    age_group = cluster_row['Age_encoded']
    income_level = cluster_row['Income_encoded']
    ctr = cluster_row['Click-Through Rates (CTR)']
    
    # Assigning cluster names based on observed characteristics
    if weekend_activity > 6 and likes_reactions > 3000 and gender == 'Male' and age_group == 25-34 and income_level == 80-100:
        return "Weekend Warriors"
    elif weekend_activity > 4.5 and likes_reactions > 3500 and gender == 'Male' and age_group == 25-34 and income_level > 100 :
        return "Engaged Professionals"
    elif weekend_activity > 3.5 and likes_reactions > 5000 and gender == 'Male' and age_group == 25-34 and income_level == 60-80 and ctr < 0.1:
        return "Low-Key Users"
    elif weekend_activity > 5.5 and likes_reactions < 4000 and gender == 'Female' and age_group == 25-34 and income_level == 60-80:
        return "Active Explorers"
    elif weekend_activity > 3 and likes_reactions < 3500 and gender == 'Female' and age_group == 25-34 and income_level == 0-20  and ctr < 0.1:
        return "Budget Browsers"
    else:
        return "Other"

# Applying the function to assign cluster names
cluster_summary['Cluster Name'] = cluster_summary.apply(assign_cluster_name, axis=1)

# Displaying the cluster names
print(cluster_summary['Cluster Name'])
```

```python
# // Question No. 9 \\

import numpy as np
import matplotlib.pyplot as plt

# Selecting numerical features for the radar chart
selected_features = ['Time Spent Online (hrs/weekday)', 'Time Spent Online (hrs/weekend)', 'Likes and Reactions', 'Click-Through Rates (CTR)', 'Conversion Rates']

# Normalizing the mean values of selected features for each cluster
normalized_means = (cluster_summary[selected_features] - cluster_summary[selected_features].min()) / (cluster_summary[selected_features].max() - cluster_summary[selected_features].min())

# Number of clusters
num_clusters = len(cluster_summary)

# Angle for each axis
angles = np.linspace(0, 2 * np.pi, len(selected_features), endpoint=False).tolist()

# Plotting radar charts for each cluster
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))
for i in range(num_clusters):
    values = normalized_means.iloc[i].tolist()
    values += values[:1]  # To close the loop
    if len(angles) != len(values):
        angles += angles[:1]  # To match the length of angles and values
    ax.plot(angles, values, linewidth=1, linestyle='solid', label=f'Cluster {i}')
    ax.fill(angles, values, alpha=0.25)
ax.set_yticklabels([])
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
plt.xticks(angles[:-1], selected_features)
plt.title('Cluster Profiles: Radar Chart')
plt.legend(loc='upper right')
plt.show()
```

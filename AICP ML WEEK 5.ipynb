```python
# // Question No. 4 \\

import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('E:\MNS-UET BS IT\AICP ML INTERNSHIP\ML Internship Task 5/Instagram-Reach.csv')

# Convert 'Date' column to datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Extract the day from the 'Date' column
df['Day'] = df['Date'].dt.day_name()

# Group by the day and calculate the total reach
daily_reach = df.groupby('Day')['Instagram reach'].sum()

# Reorder the days to match the calendar order
ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_reach = daily_reach.reindex(ordered_days)

# Plot the total reach for each day using a bar chart
plt.figure(figsize=(12, 6))
daily_reach.plot(kind='bar', color='skyblue')
plt.title('Total Instagram reach for Each Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Instagram reach')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# // Question No. 9 \\

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load the data
df = pd.read_csv('E:\MNS-UET BS IT\AICP ML INTERNSHIP\ML Internship Task 5/Instagram-Reach.csv')

# Convert 'Date' column to datetime type and set as index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Ensure data is sorted by date
df.sort_index(inplace=True)

# Decompose the time series
decomposition = seasonal_decompose(df['Instagram reach'], model='additive', period=7)  # Assuming weekly seasonality

# Plot the components
plt.figure(figsize=(12, 8))

plt.subplot(411)
plt.plot(df.index, df['Instagram reach'], label='Original')
plt.legend(loc='upper left')
plt.title('Decomposition of Instagram reach')

plt.subplot(412)
plt.plot(df.index, decomposition.trend, label='Trend')
plt.legend(loc='upper left')

plt.subplot(413)
plt.plot(df.index, decomposition.seasonal, label='Seasonality')
plt.legend(loc='upper left')

plt.subplot(414)
plt.plot(df.index, decomposition.resid, label='Residuals')
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# // Question No. 10 \\ 

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Load the data
df = pd.read_csv('E:\MNS-UET BS IT\AICP ML INTERNSHIP\ML Internship Task 5/Instagram-Reach.csv')


# Convert 'Date' column to datetime type and set as index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Ensure data is sorted by date
df.sort_index(inplace=True)

# Plot the autocorrelation function (ACF)
plt.figure(figsize=(12, 6))
plot_acf(df['Instagram reach'], lags=30, alpha=0.05)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function (ACF)')
plt.show()

# Plot the partial autocorrelation function (PACF)
plt.figure(figsize=(12, 6))
plot_pacf(df['Instagram reach'], lags=30, alpha=0.05)
plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation')
plt.title('Partial Autocorrelation Function (PACF)')
plt.show()
```

```python
# // Question No. 11 \\

import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('E:\MNS-UET BS IT\AICP ML INTERNSHIP\ML Internship Task 5/Instagram-Reach.csv')

# Convert 'Date' column to datetime type and set as index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Resample the data to the desired frequency (e.g., daily, weekly)
df = df.resample('D').mean()  # Change 'D' to the desired frequency code (e.g., 'W' for weekly)

# Ensure data is sorted by date
df.sort_index(inplace=True)

# Split the data into training and testing sets
train_size = int(len(df) * 0.8)
train_data, test_data = df[:train_size], df[train_size:]

# Define the SARIMA model
order = (1, 1, 1)  # (p, d, q)
seasonal_order = (1, 1, 1, 7)  # (P, D, Q, m)
sarima_model = SARIMAX(train_data, order=order, seasonal_order=seasonal_order, enforce_stationarity=False, enforce_invertibility=False)

# Fit the SARIMA model to the training data
sarima_result = sarima_model.fit()

# Make predictions on the testing data
predictions = sarima_result.get_forecast(steps=len(test_data))
predicted_values = predictions.predicted_mean
conf_int = predictions.conf_int()

# Evaluate the model
mse = mean_squared_error(test_data, predicted_values)
print('Mean Squared Error (MSE):', mse)

# Plot the predicted values and confidence intervals
plt.figure(figsize=(12, 6))
plt.plot(train_data.index, train_data, label='Training Data')
plt.plot(test_data.index, test_data, label='Testing Data')
plt.plot(test_data.index, predicted_values, color='red', label='Predicted Values')
plt.fill_between(test_data.index, conf_int.iloc[:, 0], conf_int.iloc[:, 1], color='pink', alpha=0.3, label='Confidence Interval')
plt.title('SARIMA Model Prediction')
plt.xlabel('Date')
plt.ylabel('Instagram reach')
plt.legend()
plt.show()
```

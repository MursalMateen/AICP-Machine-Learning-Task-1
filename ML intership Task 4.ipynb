```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np

# Load the dataset
df = pd.read_csv('C:/Users/Lenovo/Desktop/userbehaviour.csv')

# Assuming you want to cluster based on screen time and amount spent
X = df[['Average Screen Time', 'Average Spent on App (INR)']].dropna()

# K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Segment'] = kmeans.fit_predict(X)

# Check the number of segments created
print("Number of Segments:", len(np.unique(kmeans.labels_)))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting the segments
sns.scatterplot(x='Average Screen Time', y='Average Spent on App (INR)', hue='Segment', data=df, palette='viridis')
plt.title('Visualization of User Segments')
plt.xlabel('Average Screen Time (minutes)')
plt.ylabel('Average Spent on App (INR)')
plt.legend(title='User Segment')
plt.show()
```
